```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [100, 150, 200, 250, 300, 350, 00, 450, 500, 550, 600, 650]
phones = [20, 30, 40, 50, 60, 70, 80, 90, 100, 110, 120, 130]

plt.plot(months, sales , marker='o', linestyle='-', color='r')
plt.plot(months, phones , marker='s', linestyle='--', color='b')
plt.title('Monthly Sales and Phone Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid()
plt.show()



```

```python
import matplotlib.pyplot as plt

products = ['laptop', 'phone', 'tablet', 'headphones']
sales = [100, 150, 200, 250]

plt.bar(products, sales, color=['red', 'blue', 'green', 'orange'])
plt.title('Product Sales')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt

departments = ['HR', 'Finance', 'IT', 'Marketing', 'Sales']
headcount = [40, 10, 35, 25, 5]

plt.barh(departments, headcount, color=['cyan', 'magenta', 'yellow', 'blue', 'green'])
plt.title('Department Headcount')
plt.xlabel('Number of Employees')
plt.ylabel('Departments')
plt.grid()  
plt.show()

```

```python
import matplotlib.pyplot as plt

ages = [23, 45, 31, 22, 34, 28, 40, 29, 33, 27]

plt.hist(ages, bins=5, color='purple', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

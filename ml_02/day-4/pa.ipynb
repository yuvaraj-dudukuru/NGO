```python
import pandas as pd
import numpy as np

raw_data = {
    'Employee_ID': [101, 102, 102, 103, 104],
    'Name': ['Abrar', 'Haresh', 'Haresh', 'Priyanka', 'Ram Teja'],
    'Salary': ['65000', 'Missing', '70000', '85000', None],
    'Department': ['IT', 'ML', 'ML', 'HR', 'Executive']
}

df = pd.DataFrame(raw_data)

print(df.shape)
print(df.info())

print(df.duplicated().sum())
df = df.drop_duplicates()

df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

salary_median = df['Salary'].median()
df['Salary'] = df['Salary'].fillna(salary_median)

print(df)
```

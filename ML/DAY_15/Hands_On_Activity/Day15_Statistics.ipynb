# Day 15 — Statistics for Data Analysis (Hands-On Activity)

This notebook applies the full statistical toolkit — mean, median, mode, variance,
standard deviation, correlation, and outlier detection — to a single customer dataset,
and translates the results into business findings.

## Step 1 — Load the dataset

**Objective:** Create a customer dataset for statistical analysis.

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    "Customer": ["C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9", "C10"],
    "Age":      [25, 32, 28, 45, 31, 52, 29, 38, 41, 27],
    "Spending": [12000, 25000, 8000, 30000, 18000, 10000, 22000, 15000, 28000, 150000],
    "Visits":   [5, 12, 3, 15, 8, 4, 11, 7, 13, 20],
})

print(df)
```

## Step 2 — Calculate the mean

```python
print("Mean spending:", df["Spending"].mean())
```

## Step 3 — Calculate the median

The mean (31,800) exceeds the median (20,000), indicating **right skew** caused by
the large value of 150,000.

```python
print("Median spending:", df["Spending"].median())
```

## Step 4 — Calculate the mode

`Visits` is used here because spending values are all distinct. All visit counts are
also distinct in this dataset, so every value is returned; with repeated values, the
most frequent would be returned.

```python
print("Mode of Visits:", df["Visits"].mode().tolist())
```

## Step 5 — Calculate the variance

The variance is in squared rupees, which is why the standard deviation is preferred
for interpretation.

```python
print("Variance of spending:", round(df["Spending"].var(), 1))
```

## Step 6 — Calculate the standard deviation

```python
print("Standard deviation of spending:", round(df["Spending"].std(), 1))
```

## Step 7 — Generate the correlation matrix

```python
print(df[["Age", "Spending", "Visits"]].corr().round(2))
```

## Step 8 — Interpret the results

- Spending is **right-skewed** (mean above median) with one large outlier.
- Spending and visits are **strongly positively correlated** (more visits accompany
  more spending).
- Age has only a weak relationship to either spending or visits.

## Step 9 — Detect outliers (IQR rule)

```python
q1, q3 = df["Spending"].quantile(0.25), df["Spending"].quantile(0.75)
iqr = q3 - q1
outliers = df[df["Spending"] > q3 + 1.5 * iqr]

print("Spending outliers:")
print(outliers[["Customer", "Spending"]])
```

## Findings

- **Typical spending is 20,000** (the median), not the 31,800 mean, which is inflated
  by one outlier.
- **Customer C10 is an outlier**, spending 150,000 — a high-value customer warranting
  special attention.
- **Visits and spending are strongly correlated**, suggesting that increasing customer
  visits may be associated with increased spending.
- **Age shows no meaningful relationship** to spending.

On completion, every statistical measure studied today has been applied to a single
dataset and the results translated into business findings.

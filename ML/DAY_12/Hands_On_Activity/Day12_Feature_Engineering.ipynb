# Day 12 — Feature Engineering (Hands-On Activity)

Transform a raw, messy order table into a rich, validated, **dashboard-ready** dataset.

We will create 10+ engineered features using arithmetic, date, string, category, conditional, and aggregated (KPI) techniques, then validate and analyze the result.

> All names and values below are fictional sample data used for demonstration only.

## Step 1 — Load the dataset

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "OrderID": [1, 2, 3, 4, 5, 6],
    "Customer": ["asha sharma", "RAVI KUMAR", "imran khan",
                 "asha sharma", "divya rao", "RAVI KUMAR"],
    "OrderDate": ["2026-01-10", "2026-02-15", "2026-03-20",
                  "2026-04-25", "2026-05-30", "2026-06-05"],
    "Category": ["Electronics", "Grocery", "Electronics",
                 "Clothing", "Grocery", "Electronics"],
    "Units": [2, 10, 1, 5, 8, 3],
    "UnitPrice": [30000, 200, 60000, 1500, 250, 25000],
    "UnitCost": [22000, 150, 45000, 900, 180, 18000]
})
print(df.head())
```

## Step 2 — Identify transformation opportunities

Note what to create:

- **Arithmetic:** Revenue, Cost, Profit, Margin% (from prices)
- **String:** FirstName (from Customer)
- **Date:** Month, Quarter (from OrderDate)
- **Category/bin:** OrderSize
- **Conditional flag:** HighValue
- **Aggregated KPI:** customer totals

## Step 3 — Create derived (arithmetic) columns

```python
df["Revenue"] = df["Units"] * df["UnitPrice"]
df["Cost"] = df["Units"] * df["UnitCost"]
df["Profit"] = df["Revenue"] - df["Cost"]
df["Margin%"] = (df["Profit"] / df["Revenue"] * 100).round(1)
print(df[["OrderID", "Revenue", "Profit", "Margin%"]])
```

## Step 4 — Create date features

```python
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
df["Month"] = df["OrderDate"].dt.month_name()
df["Quarter"] = df["OrderDate"].dt.quarter
print(df[["OrderID", "Month", "Quarter"]])
```

## Step 5 — Create string and category features

```python
df["Customer"] = df["Customer"].str.title()                       # standardize
df["FirstName"] = df["Customer"].str.split(" ").str[0]            # extract
df["OrderSize"] = pd.cut(df["Revenue"], bins=[0, 5000, 50000, 1000000],
                         labels=["Small", "Medium", "Large"])     # bin
df["HighValue"] = np.where(df["Revenue"] >= 50000, "Yes", "No")  # conditional
print(df[["Customer", "FirstName", "OrderSize", "HighValue"]])
```

## Step 6 — Create KPI / aggregated features

```python
df["CustomerTotal"] = df.groupby("Customer")["Revenue"].transform("sum")
print(df[["Customer", "Revenue", "CustomerTotal"]])
```

## Step 7 — Validate results

```python
print("Missing values:", df.isnull().sum().sum())
print("Negative profit rows:", df[df["Profit"] < 0].shape[0])
print("Margin out of range:",
      df[(df["Margin%"] < 0) | (df["Margin%"] > 100)].shape[0])
```

## Step 8 — Analyze the transformed dataset

```python
print(df.groupby("Category")["Profit"].sum().sort_values(ascending=False))
print("\nOrders by size:")
print(df["OrderSize"].value_counts())
```

## Step 9 — Document findings

- Created **10+ engineered features**: Revenue, Cost, Profit, Margin%, Month, Quarter, FirstName, OrderSize, HighValue, CustomerTotal.
- **Electronics drives the most profit.**
- The dataset is now **dashboard-ready**: you can chart profit by month/category, the order-size mix, and customer totals.

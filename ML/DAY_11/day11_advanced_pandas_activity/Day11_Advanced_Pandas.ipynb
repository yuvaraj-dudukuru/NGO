# Day 11 — Advanced Pandas: Complete Multi-Table Analysis

A complete, end-to-end workflow on **multiple tables together**: grouping, multiple
aggregations, merging, pivot tables, crosstabs, and KPI reporting — the full advanced
pandas toolkit.

## Step 1 — Load multiple datasets

```python
import pandas as pd

customers = pd.DataFrame({
    "CustomerID": [1, 2, 3, 4],
    "Name": ["Asha", "Ravi", "Imran", "Divya"],
    "City": ["Pune", "Mumbai", "Pune", "Delhi"]
})
orders = pd.DataFrame({
    "OrderID": [101, 102, 103, 104, 105],
    "CustomerID": [1, 2, 1, 3, 4],
    "Category": ["Elec", "Groc", "Elec", "Cloth", "Groc"],
    "Amount": [25000, 1500, 30000, 9000, 2000]
})
print(customers)
print(orders)
```

## Step 2 — Perform GroupBy analysis

```python
print(orders.groupby("Category")["Amount"].sum())
```

## Step 3 — Apply multiple aggregations

```python
print(orders.groupby("Category").agg(
    Total=("Amount", "sum"),
    Average=("Amount", "mean"),
    Count=("Amount", "count")
).reset_index())
```

## Step 4 — Merge tables

```python
merged = pd.merge(orders, customers, on="CustomerID", how="left")
print(merged)
```

## Step 5 — Analyze the merged (multi-table) data

```python
print(merged.groupby("City")["Amount"].sum().sort_values(ascending=False))
```

## Step 6 — Create a pivot table

```python
pivot = pd.pivot_table(merged, index="City", columns="Category",
                       values="Amount", aggfunc="sum", fill_value=0, margins=True)
print(pivot)
```

## Step 7 — Generate a KPI report

```python
kpi = merged.groupby("City").agg(
    Total_Revenue=("Amount", "sum"),
    Orders=("OrderID", "count")
).reset_index().sort_values("Total_Revenue", ascending=False)
print(kpi)
```

## Step 8 — Crosstab analysis

```python
print(pd.crosstab(merged["City"], merged["Category"]))
```

## Step 9 — Document insights

- **Pune dominates revenue** (₹64,000 from 3 orders), driven by Electronics.
- **Electronics is the top category** (₹55,000), concentrated in Pune.
- **Mumbai and Delhi contribute little** — growth opportunities.
- **Recommendation:** Focus on Pune and Electronics; investigate Mumbai/Delhi.
